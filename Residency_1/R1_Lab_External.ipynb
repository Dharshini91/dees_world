   2. sepal width in cm

### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
def ranger(a):
    return max(a) - min(a)

for i in range(0,len(iris.select_dtypes(include=["int64","float64"]).columns)):
    print("Range of " + iris.select_dtypes(include=["int64","float64"]).columns[i] + " :")
    ranger(iris.loc[:,iris.select_dtypes(include=["int64","float64"]).columns[i]])
```

### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)
