```python
import requests
```

```python
response = requests.get('https://www.timeanddate.com/moon/phases/usa/atlanta?year=2009',headers = {"Accept-Language": "en-US, en;q=0.5"})
```

```python
from bs4 import BeautifulSoup

html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

```python
movie_containers = html_soup.find_all('div', class_ = 'tb-scroll')
type(movie_containers)
len(movie_containers)
```
