Q2.	Print all the teams which have started playing between 1930-1980.

Ans:

```python
def debut_year(begin,end):
    year_array_with_limit = laliga[laliga.Debut.between(begin,end)].loc[:,"Debut"].copy()
    for i in laliga.Debut:
        if '-' in i:
            debut_begin = i.split('-')[0]
            debut_end = i.split('-')[0][0:2] + i.split('-')[1]
        else:
            debut_begin = i
            debut_end = i
        if ((begin >= debut_begin) & (begin <= debut_end)) | ((end >= debut_begin) & (end <= debut_end)):
            year_array_with_limit = year_array_with_limit.append(pd.Series(i))
    year_array_with_limit.sort_values(inplace=True)
    return year_array_with_limit

debut = pd.Series(data=debut_year('1930','1980'))
debut.reset_index(drop=True,inplace=True)
laliga[laliga.Debut.isin(debut)].loc[:,['Team','Debut']].sort_values(by='Debut').reset_index(drop=True)
```

Q4.	Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.
		Using the same function, find the team which has maximum and minimum goal difference. (7 points)
		Goal_diff_count = GoalsFor - GoalsAgainst

Ans:

```python
def Golf_Diff_Count(temp):
    df_team = temp.loc[:,['Team','GoalsFor','GoalsAgainst']]
    df_team['GoalDiff'] = 0
    for i in df_team['Team']:
        df_team['GoalDiff'] = df_team['GoalsFor'] - df_team['GoalsAgainst']
        #df_team['GoalDiff'] = df_team['GoalDiff'].pow(2).pow(1/2)
    df_team.sort_values(by='GoalDiff',ascending=False,inplace=True)
    df_team.reset_index(drop=True,inplace=True)
    df_team = df_team.loc[:,['Team','GoalDiff']]
    return df_team
```

```python
Goal_Diff = Golf_Diff_Count(laliga)
Goal_Diff
print("The Team with maximum goal difference is " + str(np.asscalar(Goal_Diff.iloc[:,0:1].head(1).values)) 
      + " with a Goal Difference of " + str(np.asscalar(Goal_Diff.iloc[:,1:2].head(1).values)))
print("The Team with minimum goal difference is " + str(np.asscalar(Goal_Diff.iloc[:,0:1].tail(1).values)) 
      + " with a Goal Difference of " + str(np.asscalar(Goal_Diff.iloc[:,1:2].tail(1).values)))
```

Q6.Group teams based on their “Best position” and print the sum of their points for all positions (6 points)

				1				25000

Ans:

```python
laliga_Sum_of_Points_By_Best_Pos = pd.DataFrame(laliga.groupby("BestPosition")["Points"].sum(),columns=["Points"])
laliga_Sum_of_Points_By_Best_Pos.reset_index(inplace=True)
```
